```python
import numpy as np
import pylab as pl
import scipy
from scipy.spatial.distance import cdist
```

```python
np.random.seed(166)
data = np.random.randn(100, 2)
labels = np.random.randint(0,2, 100)
labels
```

```python
data.shape
ax = pl.figure(figsize = (10,10)).add_subplot(111)
ax.axis('equal')
ax.plot(data[:,0][labels == 0], data[:,1][labels == 0], 'o')
ax.plot(data[:,0][labels == 1], data[:,1][labels == 1], 's')

ax.set_xlabel('x')
ax.set_ylabel('y')
```

```python
testparticle = np.random.randn(2)
testparticle
```

```python
data.shape
ax = pl.figure(figsize = (10,10)).add_subplot(111)
ax.axis('equal')
ax.plot(data[:,0][labels == 0], data[:,1][labels == 0], 'o')
ax.plot(data[:,0][labels == 1], data[:,1][labels == 1], 's')
ax.plot(testparticle[0], testparticle[1], 'kx', ms=30)
ax.set_xlabel('x')
ax.set_ylabel('y')
```

```python
d = cdist(np.atleast_2d(testparticle), data)
```

```python
np.argmin(d)
np.min(d), d[:, np.argmin(d)]
k = 1
labeltest = labels[np.argmin(d)]
```

```python
ax = pl.figure(figsize = (10,10)).add_subplot(111)
ax.axis('equal')
colors = ['blue', 'orange']
ax.plot(data[:,0][labels == 0], data[:,1][labels == 0], 'o')
ax.plot(data[:,0][labels == 1], data[:,1][labels == 1], 's')
ax.plot(testparticle[0], testparticle[1], 'kx', ms=30, color = colors[labeltest])
ax.set_xlabel('x')
ax.set_ylabel('y')
```

```python
d[:, np.argsort(d)]
```

```python
k = 5
nns = data[np.argsort(d)][0][:k]
```

```python
nnslabel = labels[np.argsort(d)][0][:k]
```

```python
labeltest2 = 0 if (nnslabel == 0).sum()> (nnslabel == 1).sum() else 1
```

```python
ax = pl.figure(figsize = (10,10)).add_subplot(111)
ax.axis('equal')
colors = ['blue', 'orange']
ax.plot(data[:,0][labels == 0], data[:,1][labels == 0], 'o')
ax.plot(data[:,0][labels == 1], data[:,1][labels == 1], 's')
ax.plot(testparticle[0], testparticle[1], 'kx', ms=30, color = colors[labeltest2])
ax.plot(nns[:,0], nns[:,1], 'kx')
ax.set_xlabel('x')
ax.set_ylabel('y')
# pi neighbours blue
```
