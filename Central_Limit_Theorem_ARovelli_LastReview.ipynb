#Simulazione teorma de limite centrale

```python
import pylab as pl
import scipy as sp
import numpy as np
from scipy import stats
import seaborn
seaborn.set_style("darkgrid")
```

```python
np.random.seed(456)
mymean = 100
```

```python
df = mymean #chi2
p = 0.5
n = mymean / p #mean = n * P binomial
mydist = {}
#distribution shape
pl.hist(np.random.chisquare(mymean, size = 500));
pl.ylabel('N')
pl.xlabel('x')
```

```python
mysize = (np.random.rand(100) * 2000).astype(int)
mysize[mysize < 10] = 10
pl.plot(mysize, '.')
```

```python
for s in mysize:
  mydist[s] = np.random.chisquare(mymean, size = s)
ax = pl.figure().add_subplot(111)
for d in mydist.keys():
  ax.plot(d, mydist[d].mean(), '.')
ax.plot(pl.xlim(),[100,100], 'k', alpha = 0.5)
ax.set_xlabel("sample size")
ax.set_ylabel("sample mean")
```

```python
for s in mysize:
  mydist[s] = np.random.uniform(mymean, size = s)
ax = pl.figure().add_subplot(111)
for d in mydist.keys():
  ax.plot(d, mydist[d].mean(), '.')
ax.plot(pl.xlim(),[100,100], 'k', alpha = 0.5)
ax.set_xlabel("sample size")
ax.set_xlabel("sample mean")
```

```python
for s in mysize:
  mydist[s] = np.random.gamma(mymean, size = s)
ax = pl.figure().add_subplot(111)
for d in mydist.keys():
  ax.plot(d, mydist[d].mean(), '.')
ax.plot(pl.xlim(),[100,100], 'k', alpha = 0.5)
ax.set_xlabel("sample size")
ax.set_xlabel("sample mean")
```

```python
x2 = np.zeros(100)
for s in mysize:
  for i in range(100):
    mydist[s] = np.random.exponential(mymean, size = s)
    x2[i] = mydist[s].mean()
for d in mydist.keys():
  ax.plot(d, mydist[d].mean(), '.');
ax.plot(pl.xlim(),[100,100], 'k', alpha = 0.5)
ax.set_xlabel("sample size")
ax.set_xlabel("sample mean")
pl.figure(2)
pl.hist(x2)
```

```python
x1 = np.zeros(100)
for s in mysize:
  for i in range(100):
    mydist[s] = np.random.normal(mymean, size = s)
    x1[i] = mydist[s].mean()
ax = pl.figure().add_subplot(111)
for d in mydist.keys():
  ax.plot(d, mydist[d].mean(), '.');
ax.plot(pl.xlim(),[100,100], 'k', alpha = 0.5)
ax.set_xlabel("sample size")
ax.set_xlabel("sample mean")
pl.figure(2)
pl.hist(x1);
```
